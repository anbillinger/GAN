```python
import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow as tf
from tensorflow.keras import layers

noise = tf.random.normal([1, 100])
#generated_image = generator(noise, training=False)

#plt.imshow(generated_image[0, :, :, 0], cmap='gray')

gen_model = tf.keras.models.Sequential()
gen_model.add(layers.Dense(64*64*128*3, use_bias=False, input_shape=(100,)))
gen_model.add(layers.LeakyRelu())
gen_model.add(layers.Reshape((64,64,128*3)))
gen_model.add(layers.Conv2DTranspose(128*3, (5, 5), strides=(1, 1), padding='same', use_bias=False))
gen_model.add(layers.Conv2DTranspose(32*3, (5, 5), strides=(1, 1), padding='same', use_bias=False))
gen_model.add(layers.BatchNormalization())
gen_model.add(layers.Conv2DTranspose(3, (5, 5), strides=(1, 1), padding='same', use_bias=False,activation='tanh'))
print(gen_model.output_shape)

img = gen_model(noise,training=False)
img = (img-np.min(img))/(np.max(img)-np.min(img))
```
